```python
import tensorflow as tf
from tensorflow.python.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras import backend as K
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Getting number of GPUs working
print("GPUs:", len(tf.config.experimental.list_physical_devices('GPU')))
```

```python
#Data preparation
#We use weight and height data to predict gender
train_X = np.array([[161,57],
                    [181,88],
                    [190,95],
                    [156,54],
                    [162,58],
                    [170,60],
                    [179,75],
                    [178,74]])

train_X = StandardScaler().fit_transform(train_X)

train_Y = np.array([[0],
                    [1],
                    [1],
                    [0],
                    [0],
                    [0],
                    [1],
                    [1]])
                   
test_X = np.array([[165,57],
                   [194,92],
                   [185,78],
                   [157,56],
                   [162,60]])

test_X = StandardScaler().fit_transform(test_X)
```

```python
#Defining ANN
model = Sequential()
model.add(Dense(4, input_shape=(8, 2), activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
opt = tf.keras.optimizers.SGD(learning_rate=0.1)
model.compile(loss='mean_squared_error', optimizer=opt, metrics=['accuracy'])
```

```python
#We train the model
history = model.fit(train_X, train_Y, epochs=2500)
```

```python
#We make some predictions
y_pred = model.predict(test_X)
print(y_pred)
```
